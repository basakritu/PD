```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
#print(average)
```

```python
##space-time average

avg=[]
sum_f=[]
for idx in range(1,20000):
    csv_file_name = ('../simulation_data/PD_csv_B0/' + 'PD_' + '%07d.csv' % idx)
    force=np.loadtxt('../simulation_data/forceContact/' + 'force_contact_' + '%07d.dat' % idx)
    nor_f=force[:,4]
    avg_nor=np.average(nor_f)
    tan_f=force[:,5]
    tot_f=np.sum([nor_f, tan_f])
    avg_tan=np.average(tan_f)
    avg_f=np.average([avg_nor,avg_tan])
    #print(avg_f)
    avg.append(avg_f)
    sum_f.append(tot_f)
avg_all=np.array(avg)
#print((avg_all))
sum_all=np.array(sum_f)
print(len(sum_all))
average=np.average(avg_all)
```

```python
##generators above and below threshold but removed points have force 0.1% of average force 
gen_above=[]
gen_below=[]
for idx in range(1,20001):
    csv_file_name = ('../simulation_data/PD_csv_B0/' + 'PD_' + '%07d.csv' % idx)
    data1 = pd.read_csv(csv_file_name)
    data = data1.replace(-np.inf, 0)
    threshold=0.1*average/100
    #threshold=10
    #print(threshold)
    #print('len',len(data))
    th_x_above=[]
    th_y_above=[]
    d_th=(data[data.columns[0]] - data[data.columns[1]])/math.sqrt(2)
    for i in range(0,len(d_th)):
        if d_th[i]>threshold:
            th_x_above.append(data.iloc[i,0])
            th_y_above.append(data.iloc[i,1])
    thx_above=np.array(th_x_above)
    thy_above=np.array(th_y_above)
    #print(thx_above)
    #print(thy_above)
    co_above=0
    for ii in range(0,len(thx_above)):
        if (thx_above[ii]>average):
            co_above=co_above+1
    #print((co_above))
    gen_above.append(co_above)
    #print(thx)
    #print(len(gen_above))
    co_below=0
    for ii in range(0,len(thx_above)):
        if (thx_above[ii]<average):
            co_below=co_below+1
    #print(co_below)
    gen_below.append(co_below)
    
above_gen=np.array(gen_above)
#print((above_gen))
below_gen=np.array(gen_below)
#print(below_gen)
```

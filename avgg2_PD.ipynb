```python
##for thresholding a PD 


import numpy as np
import matplotlib
from matplotlib import pyplot as plt
Tp_below=[]
Tp_above=[]
Tp=[]
for i in range(2,2873):
    file=open('../Gudhi_Persistence/' + str(i)+'.txt','r')
    file2=file.read()
    file3=file2.replace('(0,','')
    file4=file3.replace('))','')
    file5=file4.replace(' (','')
    file6=file5.replace('[','')
    file7=file6.replace(']','')
    file8=file7.replace(' ','')
    file9=file8.replace('inf','0')
    data=file9.split(',')
    data1=np.asarray(data)
    birth=[]
    death=[]
    for j in range(0,len(data1)//2):
        birth.append(abs(float(data1[2*j])))
    birthh=np.array(birth)
    for k in range(0,len(data1)//2):
        death.append(abs(float(data1[2*k+1])))
    deathh=np.array(death)
    g_2=[]
    data=np.loadtxt('../G2perprt/'+'pimage'+'%05d_g2perprt.txt' %i)
    for j in range(0,len(data)):
        #print(file[:,2])
        g2=data[:,2]
        g_2.append(g2)
    G_2=np.array(g2)
    avg_g2_per=np.average(G_2)
    #print(avg_g2_per)
    index_below=[]
    index_above=[]
    index=[]
    for ii in range(0,len(birthh)):
        if (birthh[ii]<avg_g2_per):
            index_below.append(ii)
    in_dex_below=np.array(index_below)
    #print(len(in_dex))
    lf_below=[]
    lf_above=[]
    lf=[]
    for jj in range(0,len(in_dex_below)):
        life_span_below=(birthh[in_dex_below[jj]]-deathh[in_dex_below[jj]])
        lf_below.append(life_span_below)
    dif_below=np.array(lf_below)
    tot_per_below=np.sum(dif_below)
    Tp_below.append(tot_per_below)
    for ii in range(0,len(birthh)):
        if (birthh[ii]>avg_g2_per):
            index_above.append(ii)
    in_dex_above=np.array(index_above)
    #print(len(in_dex))
    lf_below=[]
    lf_above=[]
    lf=[]
    for jj in range(0,len(in_dex_above)):
        life_span_above=(birthh[in_dex_above[jj]]-deathh[in_dex_above[jj]])
        lf_above.append(life_span_above)
    dif_above=np.array(lf_above)
    tot_per_above=np.sum(dif_above)
    Tp_above.append(tot_per_above)
    lifespan=np.sum(birthh-deathh)
    Tp.append(lifespan)
TP=np.array(Tp)
TP_aboveavg=np.array(Tp_above)
TP_belowavg=np.array(Tp_below)
print(len(TP_belowavg))
print(len(TP_aboveavg))
print(len(TP))
```

```python
##for betti numbers


import numpy as np
import matplotlib
from matplotlib import pyplot as plt
b0_low=[]
b0_high=[]
b0=[]
print(average)
for i in range(2,2873):
    file=open('../Gudhi_Persistence/' + str(i)+'.txt','r')
    file2=file.read()
    file3=file2.replace('(0,','')
    file4=file3.replace('))','')
    file5=file4.replace(' (','')
    file6=file5.replace('[','')
    file7=file6.replace(']','')
    file8=file7.replace(' ','')
    file9=file8.replace('inf','0')
    data=file9.split(',')
    data1=np.asarray(data)
    birth=[]
    death=[]
    for j in range(0,len(data1)//2):
        birth.append(abs(float(data1[2*j])))
    birthh=np.array(birth)
    for k in range(0,len(data1)//2):
        death.append(abs(float(data1[2*k+1])))
    deathh=np.array(death)
    count_low=0
    count_high=0
    count=0
    for ii in range(0,len(birthh)):
        if (birthh[ii]>2*average and deathh[ii]<2*average):
            count_high=count_high+1
    #print(count)
    b0_high.append(count_high)
    for ii in range(0,len(birthh)):
        if (birthh[ii]>average and deathh[ii]<average):
            count=count+1
    #print(count)
    b0.append(count)
    for ii in range(0,len(birthh)):
        if (birthh[ii]>average/2 and deathh[ii]<average/2):
            count_low=count_low+1
    #print(count)
    b0_low.append(count_low)
b_0_low=np.array(b0_low)
b_0_high=np.array(b0_high)
b_0=np.array(b0)
#print((b_0))
#time=range(2,2873)
#plt.plot(time,b_0)
print(len(b_0_low))           
```

```python
class Node(object):
    _children = []
    def __init__(self, value, *args):
        self._value = value
        self._children = args

def sum_node(node):
    sum = 0
    if(node._value != None):
        sum+=node._value
        if(node._children != None):
            for i in range(0, len(node._children)):
                sum+=sum_node(node._children[i])
    return sum
root = Node(1, Node(4, Node(9), Node(6)), Node(7, None), Node(3, Node(2, True)), Node('a',Node(8)))
print(sum_node(root))
```
